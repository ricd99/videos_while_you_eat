```python
from pathlib import Path
import pandas as pd
import psycopg2
import os
from dotenv import load_dotenv
```

```python
base = Path.cwd()
data = base.parent.parent / "data" / "processed" / "channels_pp.csv"
df = pd.read_csv(data)
load_dotenv(override=True) # override=True to update if change in .env occurs, as jupyer caches env variables
```

```python
conn = psycopg2.connect(
    host=os.getenv("RDS_HOST"),
    port=5432,
    database="postgres",
    user="postgres",
    password=os.getenv("RDS_PW"),
    sslmode="require"
)

print("connected successfully!")
conn.close()
```
